```python
import base64
import requests # type: ignore
import json
import os
import mimetypes

# os.system("cd ../../;php artisan migrate:fresh --seed")
# os.system("mysql --database='shop_point' -e 'DELETE from commands where id > 1; DELETE FROM debts WHERE id>2;DELETE FROM waiting_stocks;DELETE FROM waiting_wallet_operations;DELETE FROM stocks;DELETE FROM refunds;'")
# os.system("mysql --database='shop_point' -e 'ALTER TABLE `commands` auto_increment = 1;ALTER TABLE `debts` auto_increment = 1;ALTER TABLE `waiting_stocks` auto_increment = 1;ALTER TABLE `waiting_wallet_operations` auto_increment = 1;ALTER TABLE `stocks` auto_increment = 1;ALTER TABLE `refunds` auto_increment = 1;'")
# os.system("mysql --database='shop_point' < stocks.sql")
# link_base = "http://credit.cofina.localhost/api/"
link_base = "https://credittg.cofinaonline.com/api/"

WARNING_COLOR = "\033[93m"
RESET_COLOR = "\033[0m"

def image_to_base64_old(image_path):
    with open(image_path, "rb") as image_file:
        base64_string = base64.b64encode(image_file.read()).decode("utf-8")
    return base64_string


def image_to_base64(image_path):
    with open(image_path, "rb") as image_file:
        image_data = image_file.read()
        mime_type, _ = mimetypes.guess_type(image_path)
        base64_string = f"data:{mime_type};base64," + base64.b64encode(
            image_data
        ).decode("utf-8")
    return base64_string


def get(
    path,
    data={},
    link_base=link_base,
    headers_sup={},
):
    headers = {"Content-Type": "application/json", "Accept": "application/json"}
    headers.update(headers_sup)
    response = requests.get(
        link_base + path, data=json.dumps(data), headers=headers, verify=False
    ).json()
    print(f"{WARNING_COLOR}{path}{RESET_COLOR}")
    return response


def post(
    path,
    data={},
    link_base=link_base,
    headers_sup={},
):
    headers = {"Content-Type": "application/json", "Accept": "application/json"}
    headers.update(headers_sup)
    response = requests.post(
        link_base + path, data=json.dumps(data), headers=headers, verify=False
    ).json()
    print(f"{WARNING_COLOR}{path}{RESET_COLOR}")
    return response


def put(
    path,
    data={},
    link_base=link_base,
    headers_sup={},
):
    headers = {"Content-Type": "application/json", "Accept": "application/json"}
    headers.update(headers_sup)
    response = requests.put(
        f"{link_base}{path}", data=json.dumps(data), headers=headers, verify=False
    ).json()
    print(f"{WARNING_COLOR}{path}{RESET_COLOR}")
    return response


response = post(
    "auth/login",
    {"email": "admin@cofinacorp.com", "password": "Coftg@20$*21ù!ad"},
)
admin = response["data"]
print(json.dumps(admin, indent=4))
```
